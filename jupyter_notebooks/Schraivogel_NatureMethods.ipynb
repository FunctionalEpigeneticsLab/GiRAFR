# Combine samples MT position

```r
rm(list = ls())
my_colors <- c("D" = 'black', 'I' = 'grey60', 
               'A' = 'darkgreen', 'T' = 'brown1', 
               'G' = 'orange', 'C'= 'royalblue')
input_dir = './GiRAFR_outputs/Schraivogel_NatureMethods/sample1/'
MT_file <- paste0(input_dir, '/all.MT.txt')
MT_sample1 <- read.table(MT_file, header = FALSE)
colnames(MT_sample1) <- c('gene name','structure','structure length', 'position','mutation','variant number')

input_dir = './GiRAFR_outputs/Schraivogel_NatureMethods/sample2/'
MT_file <- paste0(input_dir, '/all.MT.txt')
MT_sample2 <- read.table(MT_file, header = FALSE)
colnames(MT_sample2) <- c('gene name','structure','structure length', 'position','mutation','variant number')

input_dir = './GiRAFR_outputs/Schraivogel_NatureMethods/sample3/'
MT_file <- paste0(input_dir, '/all.MT.txt')
MT_sample3 <- read.table(MT_file, header = FALSE)
colnames(MT_sample3) <- c('gene name','structure','structure length', 'position','mutation','variant number')


save_pdf <- paste0('./figures_results/Schraivogel_NatureMethods/', 'Combined.all.MT_position_adjusted.pdf')
pdf(save_pdf, width = 16, height = 5)

options(repr.plot.width = 16, repr.plot.height = 5)
MT2 <- rbind(rbind(MT_sample1, MT_sample2), MT_sample3)
MT2[which(MT2$structure == 'U6'), 'position'] <- MT2[which(MT2$structure == 'U6'), 'position']-250
MT2[which(MT2$structure == 'gRNA'), 'position'] <- MT2[which(MT2$structure == 'gRNA'), 'position'] - 1
MT2[which(MT2$structure == 'Rest'),'position'] <- MT2[which(MT2$structure == 'Rest'),'position']+20
MT2$position <- factor(MT2$position, levels = seq(-20, 55, by =1))
MT2$mutation2 <- MT2$mutation
MT2$mutation2[grepl('D', MT2$mutation2)] <- 'D' 
MT2$mutation2[grepl('I', MT2$mutation2)] <- 'I'
MT2$mutation2 <- factor(MT2$mutation2, levels = c('A','T','G','C','D','I'))

ggplot(MT2, 
       aes(x = as.factor(position), fill = mutation2)) + geom_bar(aes(y = (..count..)/3))+
scale_fill_manual("mutation", values = my_colors)+xlab('position') +theme_classic() +
theme( panel.border =element_rect(colour = 'black', fill = NA, size = 1), 
      plot.title = element_text(size = 18))+scale_x_discrete(limits = levels(MT2$position))+
ggtitle(label = 'TGTGGAAAGGACGAAACACCGNNNNNNNNNNNNNNNNNNNNGTTTAAGAGCTATGCTGGAAACAGCATAGCAAGTT')


ggplot(MT2, 
       aes(x = as.factor(position), fill = mutation2)) + geom_bar(aes(y = (..count..)/3))+
scale_fill_manual("mutation", values = my_colors)+xlab('position') +theme_classic() +
theme(panel.border =element_rect(colour = 'black', fill = NA, size = 1), 
      axis.title.x=element_blank(),
      axis.line=element_blank(),
      axis.text.x=element_blank(),
      axis.ticks.x=element_blank(),
      plot.title = element_text(size = 18))+
scale_x_discrete(limits = levels(MT2$position))+scale_y_continuous(breaks = c(0,160))

dev.off()
```

# cell number comparison

```r
sample1 <- read.csv('../../public_data/TAP_seq/GSM4012694_TASC_SCREEN_chr8_sample1.pertStatus.csv')
sample2 <- read.csv('../../public_data/TAP_seq/GSM4012695_TASC_SCREEN_chr8_sample2.pertStatus.csv')
sample3 <- read.csv('../../public_data/TAP_seq/GSM4012696_TASC_SCREEN_chr8_sample3.pertStatus.csv')
```

Matrix with booleans to indicate perturbation state

```r
sample1_nr <- as.data.frame(colSums(sample1[,-1] > 0))
colnames(sample1_nr) <- 'paper'
table(sample1_nr)
#sum(table(sample1_nr)[3:length(table(sample1_nr))]) cells with > 2sgRNAs
sample2_nr <- as.data.frame(colSums(sample2[,-1] > 0))
colnames(sample2_nr) <- 'paper'
table(sample2_nr)
#sum(table(sample2_nr)[3:length(table(sample2_nr))]) cells with > 2sgRNAs
sample3_nr <- as.data.frame(colSums(sample3[,-1] > 0))
colnames(sample3_nr) <- 'paper'
table(sample3_nr)
#sum(table(sample3_nr)[3:length(table(sample3_nr))]) cells with > 2sgRNAs
```

```r
sample1_nr$cb <- unlist(lapply(rownames(sample1_nr), function(x) substr(x, 9,24)))
sample2_nr$cb <- unlist(lapply(rownames(sample2_nr), function(x) substr(x, 9,24)))
sample3_nr$cb <- unlist(lapply(rownames(sample3_nr), function(x) substr(x, 9,24)))
```

```r
filtered_barcodes <- read.table('/lustre1/project/stg_00064/projects/cropseq/267genes/develop/paper/TAP-seq/barcodes.tsv', header = FALSE)
filtered_barcodes$V1 <- sub('-1','',filtered_barcodes$V1)
nrow(filtered_barcodes)
n_col = max(count.fields('../TAP-seq/auto_pool///cells.gRNA.txt', sep = "\t"))
cells.gRNA <- read.table('../TAP-seq/auto_pool/cells.gRNA.txt', 
                         header = FALSE, fill = TRUE, stringsAsFactors = FALSE,col.names =1:n_col)
                                     
colnames(cells.gRNA)[1:7]= c('cb','feature_order','target','WT','num_UMI_WT', 'MT','num_UMI_MT')
cells.num_features <- cells.gRNA %>% group_by(cb) %>% summarise(num_features = max(feature_order))
nrow(cells.num_features)
table(cells.num_features$num_features)
protospacer_calls_per_cell <-read.csv('/lustre1/project/stg_00064/projects/cropseq/267genes/develop/public_data/TAP_seq/feature_barcoding_analysis/TAP_seq/outs/crispr_analysis/protospacer_calls_per_cell.csv', stringsAsFactors = FALSE)
protospacer_calls_per_cell$cell_barcode <- sub('-1','',protospacer_calls_per_cell$cell_barcode )
sum(sample1_nr$cb %in% filtered_barcodes$V1)
```

Only take overlappped barcodes for comparison

```r
filtered_barcodes.merged <- merge(filtered_barcodes, sample1_nr, by.x = 'V1', by.y = 'cb')
filtered_barcodes.merged <- merge(filtered_barcodes.merged, cells.num_features, by.x = 'V1', by.y = 'cb', all.x = TRUE)
filtered_barcodes.merged <- merge(filtered_barcodes.merged,protospacer_calls_per_cell, suffixes = c('.girafr','.cellranger'),
                                  by.x = 'V1', by.y ='cell_barcode', all.x = TRUE)
filtered_barcodes.merged[is.na(filtered_barcodes.merged)] <- 0
```

```r
filtered_barcodes_merged.all.long <-  gather(filtered_barcodes.merged[,1:4], key = 'method', value = 'num_features', -c(V1))
```

```r
ggplot(filtered_barcodes_merged.all.long, aes(x=num_features, fill = method))+geom_bar(position="dodge")+
theme_bw()
```
